Sea la ecuación logística:

$$
X_{t+1} = rX_t(1 - X_t)
$$

A continuación sobre esta ecuación logistica vamos a calcular y graficar 48 periodos de tiempo para 9 configuraciones diferentes de valores iniciales $X_0$ y $r$

```python
# Importamos paquetes requeridos
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ecuacion_logistica(r, x_0, t):
    serie = [x_0]

    for i in range(t - 1):
        x_1 = r*serie[-1] * (1 - serie[-1])
        serie.append(x_1)
    return serie

def graficadora(serie_input):
    plt.plot(serie_input)
    plt.xlabel('Tiempo')
    plt.ylabel('Población')
    plt.grid(True)
    plt.ylim(-0.1,1)
```

```python
plt.figure(figsize=(15,10))
plt.suptitle("series de tiempo ecuación logistica", size=20)

plt.subplot(4,3,1)

serie_a = ecuacion_logistica(0.25,0.3,48)
graficadora(serie_a)

plt.subplot(4,3,2)

serie_b = ecuacion_logistica(0.75,0.3,48)
graficadora(serie_b)

plt.subplot(4,3,3)

serie_c = ecuacion_logistica(1.25,0.3,48)
graficadora(serie_c)

plt.subplot(4,3,4)

serie_d = ecuacion_logistica(2,0.3,48)
graficadora(serie_d)

plt.subplot(4,3,5)

serie_e = ecuacion_logistica(3.2,0.3,48)
graficadora(serie_e)

plt.subplot(4,3,6)

serie_f = ecuacion_logistica(3.54,0.3,48)
graficadora(serie_f)

plt.subplot(4,3,7)

serie_g = ecuacion_logistica(3.99,0.3,48)
graficadora(serie_g)

plt.subplot(4,3,8)

serie_h = ecuacion_logistica(3.99,0.301,48)
graficadora(serie_h)

plt.subplot(4,3,9)

serie_i = ecuacion_logistica(3.99,0.3000001,48)
graficadora(serie_i)

plt.subplot(4,3,10)

graficadora(serie_g)
graficadora(serie_h)

plt.subplot(4,3,11)

graficadora(serie_g)
graficadora(serie_i)

plt.subplots_adjust(hspace=0.5)
plt.show()

```
