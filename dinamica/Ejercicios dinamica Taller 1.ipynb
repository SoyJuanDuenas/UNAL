```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def serie_med(d0, lamda,t):
    serie = [d0]

    for i in range(t - 1):
        x_1 = serie[-1] - serie[-1]*lamda + d0
        serie.append(x_1)
    return serie
def graficadora(serie_input):
    plt.plot(serie_input)
    #plt.xlabel('Tiempo')
    #plt.ylabel('Poblaci贸n')
    plt.grid(True)
    #plt.ylim(-0.1,1)
    plt.show()
```

```python
serie = serie_med(2,0.25,24)
graficadora(serie)
```

## Tercer punto

```python
def serie_a(d0, t):
    serie = [d0]
    for i in range(1, t):
        x_1 = d0 * (3 ** (i * (i - 1) / 2))
        serie.append(x_1)
    print(serie)
    return serie


def serie_b(d0,t):
    serie = [d0]

    for i in range(t - 1):
        x_1 = (d0 - 4)*(2**-int(i)) + 4
        serie.append(x_1)
    print(serie)
    return serie
```

```python
serie_a = serie_a(1,24)
graficadora(serie_a)
serie_b = serie_b(1,24)
graficadora(serie_b)
```

## cuarto punto 

```python
def serie_cuarto(x0,x1,t):
    serie = [x0,x1]
    
    for i in range(t-1):
        x_2 = 1 + 3*serie[-1] - 2*serie[-2]   
        serie.append(x_2)
    print(serie)
    return serie
```

```python
serie_cuarto = serie_cuarto(5,3,24)
graficadora(serie_cuarto)
```
