```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def serie_med(d0, lamda,t):
    serie = [d0]

    for i in range(t - 1):
        x_1 = serie[-1] - serie[-1]*lamda + d0
        serie.append(x_1)
    return serie
def graficadora(serie_input):
    plt.plot(serie_input)
    plt.xlabel('Tiempo')
    plt.ylabel('Poblaci贸n')
    plt.grid(True)
    #plt.ylim(-0.1,1)
    plt.show()
```

```python
serie = serie_med(2,0.25,24)
graficadora(serie)
```
